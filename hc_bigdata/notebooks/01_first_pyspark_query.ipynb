# First PySpark Notebook — On-Premise Setup Verification

This notebook proves the local Spark setup works end-to-end:
Java 11 + Spark 3.5.5 + winutils.exe + PySpark + JupyterLab, all running on this laptop.

Equivalent to a first Databricks notebook cell reading a table and running a query.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("hr_bigdata_first_query") \
    .master("local[*]") \
    .getOrCreate()

spark
```

```python
df = spark.read.csv("../data/employees.csv", header=True, inferSchema=True)
df.show()
```

```python
df.createOrReplaceTempView("employees")

spark.sql("""
    SELECT department, COUNT(*) AS headcount, ROUND(AVG(salary), 2) AS avg_salary
    FROM employees
    GROUP BY department
    ORDER BY department
""").show()
```

```python
spark.stop()
```
