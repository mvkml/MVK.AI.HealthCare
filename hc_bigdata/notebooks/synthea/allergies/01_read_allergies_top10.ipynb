# Step 1 — Read `allergies.csv` and show top 10 records

First step of the Synthea CSV → SQL Server pipeline, for learning purposes.
This notebook only reads the file and displays data — no writing anywhere yet.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("synthea_allergies_read") \
    .master("local[*]") \
    .getOrCreate()

spark
```

```python
allergies_path = "../../../data/patient_details/synthea/allergies.csv"

# df = spark.read.csv(allergies_path, header=True, inferSchema=True)
df = spark.read.option("header", "true").csv(allergies_path)
print("Row count:", df.count())
print("Columns:", df.columns)
```

```python
df.show(10, truncate=False)
```

```python
df.printSchema()
```
